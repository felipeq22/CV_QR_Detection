### 1.Overview

In order to train a model to extract the QR Codes of images we used two approaches. These were done with the objective of analyzing how using a pre trained model vs builting one from the scratch can allow us to make better predictions. Thus, we are going to compare them and analyze their results. Both models were trained to make a regression prediction of the bounding boxes of the QR codes.

1.   Creating a deep learning model from scratch.
2.   Fine tuning a pre trained model.

In order to train the models we used a data that was obtained from Roboflow QR Code data set at the following link https://universe.roboflow.com/lihang-xu/qr-code-oerhe/browse?queryText=&pageSize=50&startingIndex=0&browseQuery=true.
Roboflow provided us with 502 and 60 images for training and testing, respectively.

After training both models, we created a local API-.exe file in order to read images given by the user and extract the QR codes.

### **8. Discussion**

This approach of selecting two models was in order to detect which kind of models adapt better to certain circustances like availibility of data, computing resources and time. We can clearly see from the results that the regular CNN architecture predicts much better than the fine tuned model which is paradogically since the VGG is a more robust model. This could be for certain factors like the number of epochs that was trained. Also if a CNN architecture added at the end of the VGG is good enough to fine tune a model.

Now, we see that the CNN got decent results but there seems to be certain lack of predictability since all the bounding boxes predicted have the same value. This could be because of various reasons. For example the complexity of the model, the data set is too small or it may be overfitted. Thus, for future work we must get in depth about this issues and try to get a better performance model
